```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Exercise 2

```python
uniform_arr = np.random.uniform(0,1,1000)
normal_1_arr = np.random.normal(0,1,1000)
normal_2_arr = np.random.normal(0,10,1000)
beta_1_arr = np.random.beta(1,1,1000)
beta_2_arr = np.random.beta(2,2,1000)
beta_3_arr = np.random.beta(3,9,1000)

mean = np.array([0, 0])
cov1 = np.array([[1, 0], [0, 1]])
cov2 = np.array([[2, 1], [1, 2]])
multi_normal_1_arr = np.random.multivariate_normal(mean, cov1, 10000)
multi_normal_2_arr = np.random.multivariate_normal(mean, cov2, 10000)
```

Exercise 3

```python
plt.hist(uniform_arr, bins=30, density=True, label='Uniform Distribution on interval [0,1]')

# 1/(1-0)
x = np.linspace(0, 1, 1000)
plt.plot(x, np.ones_like(x), label='Analytical Formula')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.legend(loc='upper left')

plt.show()
```

```python
plt.hist(normal_1_arr, bins=30, density=True, label='Normal Distribution with mean = 0 and std = 1')

x = np.linspace(-3, 3, 1000)
y = np.exp(-0.5*x**2) / np.sqrt(2*np.pi)
#  e^((x  )^2/2^2) /*sqrt(2)
plt.plot(x, y, label='Analytical Formula')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend(loc='upper left')

plt.show()
```

```python
plt.hist(normal_2_arr, bins=30, density=True, label='Normal Distribution with mean = 0 and std = 10')

x = np.linspace(-30, 30, 1000)
y = np.exp(-0.5*(x**2/100)) / (10 * np.sqrt(2*np.pi))
plt.plot(x, y, label='Analytical Formula')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend(loc='upper left')

plt.show()
```

```python
plt.hist(beta_1_arr, bins=30, density=True, label='Beta Distribution with parameters (1,1)')

x = np.linspace(0, 1, 1000)
def beta_pdf(x, a, b):
    return x**(a-1) * (1-x)**(b-1) / (np.math.gamma(a) * np.math.gamma(b) / np.math.gamma(a+b))

plt.plot(x, beta_pdf(x, 1, 1), label='Analytical Formula')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.legend(loc='upper left')

plt.show()
```

```python
plt.hist(beta_2_arr, bins=30, density=True, label='Beta Distribution with parameters (2,2)')

x = np.linspace(0, 1, 1000)
plt.plot(x, beta_pdf(x, 2, 2), label='Analytical Formula')


# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.legend(loc='upper left')

plt.show()
```

```python
plt.hist(beta_3_arr, bins=30, density=True, label='Beta Distribution with parameters (3,9)')

x = np.linspace(0, 1, 1000)
plt.plot(x, beta_pdf(x, 3, 9), label='Analytical Formula')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.legend(loc='upper left')

plt.show()
```

```python
# Set up parameters
N = 1000
a = np.random.normal(loc=2, scale=0.1, size=N)
b = np.random.normal(loc=0.2, scale=0.2, size=N)
x = np.random.normal(loc=0, scale=3, size=N)
sigma = abs(np.random.normal(loc=0, scale=1, size=N))

# Compute u
u = a + x*b

# Sample y from normal distribution
y = np.random.normal(loc=u, scale=sigma)

# Plot y as a function of x
fig, ax = plt.subplots()
ax.scatter(x, y, s=5)

# Plot lines of y=a+b*x for randomly sampled a and b pairs
for i in range(100):
    a_i = np.random.choice(a)
    b_i = np.random.choice(b)
    x_i = np.linspace(-6, 6, 100)
    y_i = a_i + b_i*x_i
    ax.plot(x_i, y_i, color='green', linewidth=1, alpha=0.5)

# Set plot labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Comparison of y values and lines of y=a+b*x')

# Show plot
plt.show()
```
